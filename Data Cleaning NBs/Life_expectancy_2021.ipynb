```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

LE = pd.read_excel('C:/Users/gonca/Downloads/tps00205_spreadsheet.xlsx', sheet_name='Sheet 1',  skiprows = 9,  nrows= 53, usecols = 'A:Y')
print(LE)
```

```python
LE = LE.dropna(axis=1, thresh=15) 
```

```python
LE = LE.tail(-1)
LE
```

```python
LE.replace(to_replace = ':', value = 0 , inplace = True)
LE
```

```python
LE.rename(columns = {'TIME':'Countries'}, inplace = True)
```

```python
LE.dtypes
```

```python
#LE =LE[(LE != 0).all(1)]
```

```python
tmp = LE.copy()    
tmp.sort_values(by="2021", ascending=False, inplace=True)

plt.figure(figsize = (8, 10))
ax = sns.barplot(data=tmp, x="2021", y="Countries", color="skyblue")
ax.bar_label(ax.containers[0])
plt.title(f"Life expectancy at birth in 2021")
plt.show()
```
