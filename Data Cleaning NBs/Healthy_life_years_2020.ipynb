```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

HLY = pd.read_excel('C:/Users/gonca/Downloads/hlth_hlye_spreadsheet.xlsx', sheet_name='Sheet 1',  skiprows = 9,  nrows= 34, usecols = 'A:AI')
print(HLY)
```

```python
HLY = HLY.dropna(axis=1, thresh=15) 
```

```python
HLY = HLY.tail(-1)
HLY
```

```python
HLY.replace(to_replace = ':', value = 0, inplace = True)
HLY
```

```python
HLY.rename(columns = {'TIME':'Countries'}, inplace = True)
```

```python
tmp = HLY.copy()
tmp.sort_values(by="2020", ascending=False, inplace=True)

plt.figure(figsize = (8, 10))
ax = sns.barplot(data=tmp, x="2020", y="Countries", color="skyblue")
ax.bar_label(ax.containers[0])
plt.title(f"Healthy life years in absolute value at birth in 2020")
plt.show()
```
